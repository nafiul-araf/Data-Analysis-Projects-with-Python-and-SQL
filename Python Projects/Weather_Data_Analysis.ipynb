**Q. 16) What is the average temperature for each month?**

```python
df.dtypes
```

```python
df['Date/Time']=pd.to_datetime(df['Date/Time'])
df.dtypes
```

```python
df['Year']=df['Date/Time'].dt.year
df['Month']=df['Date/Time'].dt.month
df.head()
```

```python
import calendar
```

```python
df['Month']=df['Month'].apply(lambda x: calendar.month_abbr[x])
df.head()
```

```python
df_avg_temp=df.groupby('Month')['Temp_C'].mean().reset_index(name='Average Temperature')

sns.lineplot(x='Month', y='Average Temperature', data=df_avg_temp)

plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.title('Average Temperature by Month')

plt.show()
```

**Q. 17) How does the temperature relate to the dew point temperature across different months?**

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(data=df, x='Temp_C', y='Dew Point Temp_C', hue='Month', size=100)

plt.xlabel('Temperature (°C)')
plt.ylabel('Dew Point Temperature (°C)')
plt.title('Temperature vs. Dew Point Temperature by Month')

legend=plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

**Q. 18) What are the details of the dataset for the top 5 most frequent weather conditions?**

```python
top_5_weather_conditions=df['Weather Condition'].value_counts().head(5).index
top_5_weather_conditions
```

```python
df_top_5_weather=df[df['Weather Condition'].isin(top_5_weather_conditions)]
df_top_5_weather
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_top_5_weather, x='Weather Condition', hue='Month')

plt.ylabel('Count')
plt.title('Frequency of top 5 Weather Condition with Month')
legend=plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```
