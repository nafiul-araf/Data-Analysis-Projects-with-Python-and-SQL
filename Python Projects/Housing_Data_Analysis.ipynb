**Q. 7) Show the total count of records of each area, where average price is less than 10000**

```python
ap_less_100000=df[df['average_price'] < 100000]

area_count=ap_less_100000.groupby('area')['area'].count().reset_index(name='count')
area_count
```

```python
plt.figure(figsize=(15, 10))

sns.barplot(data=area_count, x='area', y='count', palette='deep')
plt.xlabel('', fontsize=14)
plt.ylabel('Count of Records', fontsize=14)
plt.title('Count of Records for Each Area (Average Price < 10000)', fontsize=16)
plt.xticks(rotation=90, ha='center')
plt.tight_layout()

plt.show()
```

```python
month_map={1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
           7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}

df['month']=df['month'].map(month_map)
df.head()
```

```python
plt.figure(figsize=(15, 8))

sns.scatterplot(data=df, x='average_price', y='houses_sold', hue='month', size='no_of_crimes', sizes=(10, 100), legend='full')
handles, labels=plt.gca().get_legend_handles_labels()  # Get current handles and labels
handles=handles[:len(df['month'].unique())]  # Keep only the handles for 'month'
labels=labels[:len(df['month'].unique())]    # Keep only the labels for 'month'
plt.legend(handles, labels)  # Update the legend to display only 'Month'
plt.title('Scatter Plot of Houses Sold vs Average Price, Colored by Month')
plt.show()
```
