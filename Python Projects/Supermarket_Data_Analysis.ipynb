# Bivariate and Multivariate Analysis

```python
df.head()
```

**1. Does COGS affetct the ratings that the customers provide?**

```python
sns.scatterplot(data=df, x='Rating', y='cogs', hue='Rating')
plt.legend(title='Rating', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

**1. Does Gross Income affetct the ratings that the customers provide?**

```python
sns.scatterplot(data=df, x='Rating', y='gross income');
```

**3. Find most profitable branch as per gross income**

```python
df.groupby('Branch')['gross income'].sum().sort_values(ascending=False)
```

**4. Is there any relationship between Gender and Gross Income?**

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Gender', y='gross income', hue='City', data=df)
plt.legend(title='Rating', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

**5. Find the product line that generates most income**

```python
sns.barplot(x='Product line', y='gross income', hue='Product line', data=df)
plt.legend([], [], frameon=False)
plt.xticks(rotation=90)
plt.show()
```

**6. Find highest Unit Price in the product line**

```python
df.groupby('Product line')['Unit price'].max().head(1).index[0]
```

**7. Find different payment method by customers citywise**

```python
payment_counts=df.groupby(['City', 'Payment'])['Payment'].count().reset_index(name='Count')

payment_counts_sorted=payment_counts.sort_values(by='Count', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Count', hue='Payment', data=payment_counts_sorted)

plt.legend(title='Rating', loc='upper left', bbox_to_anchor=(1, 1))


plt.xlabel('Count')
plt.ylabel('City')
plt.title('Payment Methods by City')

plt.show()
```

**8. Which product line is purchased in high quantity?**

```python
df.groupby('Product line')['Quantity'].sum().sort_values(ascending=False).head(1).index[0]
```

**9. Display daily sales by day of week**

```python
df['dow']=df['Date'].dt.dayofweek
df.head()
```

```python
sns.lineplot(data=df, x='dow', y='Total');
```

```python
df['dow']=df['dow'].map({
    0: 'Mon',
    1: 'Tue',
    2: 'Wed',
    3: 'Thu',
    4: 'Fri',
    5: 'Sat',
    6: 'Sun'
})
```

```python
sns.lineplot(data=df, x='dow', y='Total');
```

**10. Which will be the highest months for sales?**

```python
df['Month']=df['Date'].dt.month
df.head()
```

```python
df.Month.value_counts()
```

```python
df['Month']=df['Month'].map({
    0: 'Jan', 1: 'Feb', 2: 'Mar', 3: 'Apr', 4: 'May', 5: 'Jun',
    6: 'Jul', 7: 'Aug', 8: 'Sep', 9: 'Oct', 10: 'Nov', 11: 'Dec'
})


sns.barplot(x='Month', y='Total', hue='Month', data=df)
plt.legend([], [], frameon=False)
plt.xticks(rotation=90)
plt.show()
```
